Считываем данные из файла
Если файл состоит из одной строки:

```python

f = open('file_name.txt')
s = f.readline().strip() # В переменную s считали строку
f.close() # Не забываем закрывать файл!
```

Если файл состоит из любого количества строк:
Вариант 1

```python
f = open('file_name.txt')
a = [] # Список, в котором будем хранить строки
for s in f:
    s = s.strip() # Удаляем '\n'
    a.append(s)
f.close()
```

Вариант 2

```python
f = open('file_name.txt')
a = [s.strip() for s in f]
f.close()
```

Например, для файла
100 ЕГЭ
Информатика
100 баллов

В a окажется список строк ['PRO100 ЕГЭ', 'Информатика', '100 баллов'].

Если файл состоит из любого количества строк, в каждой из которых записано одно число:

Вариант 1

```python
f = open('file_name.txt')
a = [] # Список, в котором будем хранить числа
for s in f:
    chislo = int(s) # Переводим строку в число. Можно обойтись без strip.
    a.append(chislo)
f.close()
```

Вариант 2

```python
f = open('file_name.txt')
a = [int(s) for s in f]
f.close()
```

Например, для файла

```python
1212
2
68
8
```

В a окажется список чисел [1212, 2, 68, 8].

Если файл состоит из любого количества строк, в каждой из которых записано одно число, при этом число в первой строке означает количество чисел в файле, кроме данного числа:

Вариант 1

```python
f = open('file_name.txt')
n = int(f.readline())
a = []
for s in f:
    chislo = int(s)
    a.append(chislo)
f.close()
print(a)
```

```python
f = open('file_name.txt')
n = int(f.readline())
a = [int(s) for s in f]
f.close()
```
